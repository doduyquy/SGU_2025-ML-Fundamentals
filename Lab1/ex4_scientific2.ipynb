## 1. Challenge:
- Display 2 chart: $sin(x), cos(x)$ | $x \in \left[-\pi, \pi\right]$

## 2. Solution:

#### Import libs

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Generate data for function

<div align="center">

$$np.linspace(start, stop, num, endpoint=True)$$

</div>

---

üìù **Description**: Generates a 1D NumPy array with evenly spaced numbers.

üîß **Parameters**
| Parameter   | Type   | Description |
|-------------|--------|-------------|
| `start`     | float  | Starting value of the interval |
| `stop`      | float  | Ending value of the interval |
| `num`       | int    | Number of samples(array length) |
| `endpoint`  | bool   | If `True` (default): include `stop` as the last value &rarr; `[start, stop]` <br> If `False`: exclude `stop` &rarr; `[start, stop)` |


üìä **Formula for Spacing**

The distance between two consecutive values ($\Delta$) is:

$$
\Delta = \frac{\text{stop} - \text{start}}{\text{num} - 1}
$$

Example with our values:

$$
\Delta = \frac{\pi - (-\pi)}{265 - 1} = \frac{2\pi}{264}
$$


---


```python
# Create a domain(range) of x
num_points = 265
X = np.linspace(-np.pi, np.pi, num_points, endpoint=True)
# Calculate cos, sin for each point - x in X domain
S, C = np.sin(X), np.cos(X)
```

```python
# Check data which is generated
print(f"- X: |X|={len(X):d}, [{min(X):.2f}, {max(X):.2f}]: 10 samples [{X[0:10]}]")
print(f"- Sin(X): |S|={len(S):d}, [{min(S):.2f}, {max(S):.2f}]: 10 samples [{S[-10:-1]}]")   
print(f"- Cos(X): |C|={len(C):d}, [{min(C):.2f}, {max(C):.2f}]: 10 samples [{C[0:10]}]")
```

#### Render chart:

<div align="center">

$$matplotlib.pyplot.plot(*args, **kwargs)$$

</div>

---

üìù **Description**: draw line plot between x, y point

üîß **Parameters**

| Parameter   | Type        | Description |
|-------------|-------------|-------------|
| `x`         | array-like  | (Optional) x-coordinates of points. If not given, uses index `[0, 1, 2, ...]` |
| `y`         | array-like  | y-coordinates of points |
| `fmt`       | str         | (Optional) Format string specifying color, marker, and line style (e.g. `"ro--"` for red, circle markers, dashed line) |
| `label`     | str         | (Optional) Label for the line (used with `plt.legend()`) |
| `**kwargs`    | keyword args| Additional properties like `color`, `linewidth`, `linestyle`, `marker`, etc. |

üé® **Common `kwargs`**

| Argument      | Description | Example |
|---------------|-------------|---------|
| `color`       | Line color | `"blue"`, `"#FF0000"`, `"r"` |
| `linewidth`   | Line thickness | `2.5` |
| `linestyle`   | Line style | `"-"` (solid), `"--"` (dashed), `"-."`, `":"` |
| `marker`      | Marker style | `"o"` (circle), `"s"` (square), `"x"`, `"^"` (triangle) |

---

```python
# Show matplotlib in current output, not a new window
%matplotlib inline 

# Visualize data
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")

# Limit on x axis
# 1.1: expend limit of x, y axis to 10% from the original data(both 2 side)
expend_percent = 0.1
plt.xlim(X.min() * (1 + expend_percent), X.max() * (1 + expend_percent))
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi], 
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
# Limit on x axis   
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$',])

plt.show()
```

- Another display

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Plot sin & cos
ax.plot(X, C, color="blue", linewidth=2.0, label="cos(x)")
ax.plot(X, S, color="red", linewidth=2.0, label="sin(x)")

# Set limits
ax.set_xlim(X.min() * 1.1, X.max() * 1.1)
ax.set_ylim(-1.1, 1.1)

# Customize ticks with LaTeX
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels([r'$-1$', r'$0$', r'$+1$'])

# Highlight intersection points
ax.scatter([0], [0], color='black', zorder=5)  # origin
ax.scatter([np.pi/2, -np.pi/2], [0, 0], color='red', zorder=5)  # sin zeros
ax.scatter([0, np.pi, -np.pi], [1, -1, -1], color='blue', zorder=5)  # cos extremes

# Move spines to center
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['left'].set_position('center')


# Add grid (optional)
ax.grid(alpha=0.3, linestyle="--")

# Add legend
ax.legend(loc='upper right')

plt.show()
```
