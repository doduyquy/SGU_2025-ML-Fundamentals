## 1. Challenge:
Random display $N=50$ circles:
- Center $c_i = (x_i, y_i)$ random with $x_i, y_i \in \left [0 , 1 \right]$ 
- Color $color_i$ random with $color_i \in \left[0 , 1\right]$
- Area $area_i = \pi * r_i^2$ with random generation $r_i \in \left[0, 1\right]$ 

## 2. Solution:

#### Import libs

```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display     # Render rich output
```

#### Generate random circles

```python
%time
# Generate randomly the center (x, y) of the circle, color, area
N = 50
# rand(N): generate N numbers from [0; 1)
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

# Radius of circle - [0; 15)
pixel_point = 15
area = np.pi * (pixel_point * np.random.rand(N)) ** 2 
```

```python
# Check data which is generated
print(f"- x: len={len(x):d}, [{min(x):.2f}, {max(x):.2f}]: 10 samples [{x[0:10]}]")
print(f"- y: len={len(y):d}, [{min(y):.2f}, {max(y):.2f}]: 10 samples [{y[-10:-1]}]")   
print(f"- colors: len={len(colors):d}, [{min(colors):.2f}, {max(colors):.2f}]: 10 samples [{colors[0:10]}]")
print(f"- area: len={len(area):d}, [{min(area):.2f}, {max(area):.2f}]: 10 samples [{area[-10:-1]}]")   
```

#### Render chart:

```python
# Show matplotlib in current output, not a new window
%matplotlib inline 

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
